```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, fixed
import tkinter as tk
from tkinter import ttk
from tkinter.ttk import *

df = pd.read_csv('recomendacion.csv', index_col=0)


print(df)
```

```python
root=tk.Tk()
root.title('Modificación de recomendaciones')

# Crear widget de selección de usuario
#Map de JS 
user_select = ttk.Combobox(root, values= [str(x) for x in df.index.values])
user_select.pack(padx=10, pady=10)


results = [0, 0, 0, 0, 0];


    
    
#Función para actualizar el diccionario de valores modificados
def set_slider_values(results):
    
    global pizza_slider
    pizza_slider.set(results[0])
    global pasta_slider
    pasta_slider.set(results[1])
    global hamburguesa_slider
    hamburguesa_slider.set(results[2])
    global carne_slider
    carne_slider.set(results[3])
    global pollo_slider
    pollo_slider.set(results[4])
    
#Funcion inicial para coger los valores del arreglo    
user = ""

def set_sliders():
    
    #Get selected user values
    global user
    user = user_select.get()
    
    #Global results python way
    global results
    
    #pasa a un mapa todos los valores de la lista
    results = list(df.loc[user].values)
    
    set_slider_values(results)


#Set in a dictionary the new values stablished  
new_values = [0, 0, 0, 0, 0]

#Poner dentro del arreglo new_values los valores actualizados de los sliders
def new_item_values():
    #global new_vaues
    global pizza_slider
    new_values[0] = pizza_slider.get()
    global pasta_slider
    new_values[1] = pasta_slider.get()
    global hamburguesa_slider
    new_values[2] = hamburguesa_slider.get()
    global carne_slider
    new_values[3] = carne_slider.get()
    global pollo_slider
    new_values[4] = pollo_slider.get()
    return new_values

def set_new_dataframe():
    new_item_values()
    print(list(df.iloc[user, 1]))
    print(new_values)
    
select_button = tk.Button(root, text='Seleccionar')
select_button.pack(padx=10, pady=10)
select_button.configure(command=set_sliders)

#Sliders
pizza_label = tk.Label(root, text="Que tanto te gusta la pizza?")   
pizza_slider = tk.Scale(root, from_=0, to=5, resolution=1, orient='horizontal')
pasta_label = tk.Label(root, text="Que tanto te gusta la pasta?")   
pasta_slider = tk.Scale(root, from_=0, to=5, resolution=1, orient='horizontal')
hamburguesa_label = tk.Label(root, text="Que tanto te gusta la hamburguesa?")
hamburguesa_slider = tk.Scale(root, from_=0, to=5, resolution=1, orient='horizontal')
carne_label = tk.Label(root, text="Que tanto te gusta la carne?")                 
carne_slider = tk.Scale(root, from_=0, to=5, resolution=1, orient='horizontal')
pollo_label = tk.Label(root, text="Que tanto te gusta la pollo?")   
pollo_slider = tk.Scale(root, from_=0, to=5, resolution=1, orient='horizontal')

# Empaquetar sliders en la ventana para que hagan display
pizza_label.pack()
pizza_slider.pack()
pasta_label.pack()
pasta_slider.pack()
hamburguesa_label.pack()
hamburguesa_slider.pack()
carne_label.pack()
carne_slider.pack()
pollo_label.pack()
pollo_slider.pack()

# Crear botón de guardar cambios
save_button = tk.Button(root, text='Guardar cambios')
save_button.pack(padx=10, pady=10)

# Conectar botón de guardar cambios a la función de actualización de valores
save_button.configure(command=set_new_dataframe)

# Ejecutar bucle principal de la ventana
root.mainloop()


```
